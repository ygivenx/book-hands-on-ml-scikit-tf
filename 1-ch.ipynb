```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

import sklearn.linear_model
import sklearn.neighbors
```

```python
gdp = pd.read_csv("gdp-2019.csv")
happiness = pd.read_csv("happiness-2020.csv")
```

```python
gdp.columns = ["country", "percapita"]
```

```python
happiness.columns = ["country", "happiness"]
```

```python
data = gdp.merge(happiness)
```

```python
data.dropna(inplace=True)
```

```python
data.plot(kind='scatter', x='percapita', y='happiness')        
```

```python
model = sklearn.linear_model.Ridge(alpha=.5)
# model = sklearn.neighbors.KNeighborsRegressor(n_neighbors=3)
model.fit(np.c_[data['percapita']], data['happiness'])
data.plot(kind='scatter', x='percapita', y="happiness")
plt.plot(data['percapita'], model.predict(np.c_[data['percapita']]), color='tab:red', linewidth=3)
```
